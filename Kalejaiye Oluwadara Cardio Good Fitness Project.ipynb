## Exploratory Data Analysis

### Univariate analysis of the Numerical Variables

```python
#Creating a function that takes a numerical variable as input and returns a histogram and boxplot visualization.
def hist_box(var):
    plt.figure(figsize=(8, 5))
    sns.boxplot(var)#plots a simple boxplot
    plt.figure(figsize=(10,7))
    plt.hist(var, color='g')#plots a histogram
    plt.axvline(var.mean(), color='r', linewidth=2)#Plot mean on histogram
    plt.axvline(var.median(), color='b', linestyle='dashed', linewidth=2)#Plot median on histogram
    plt.axvline(var.mode()[0], color='w', linestyle='dashed', linewidth=2)#Plot mode on histogram
    
```

### Observations on Age

```python
hist_box(data['Age'])
```

**Observations**
* The age distribution is skewed more to right.
* Outliers exist with customers over 45 years of age.
* The majority of customers are in their 20's with the highest proportion in their mid to late twenties.

### Observations on Education

## Conclusions and Recommendations

## Conclusion

We have analyzed data for customers of the treadmill products of a retail store called Cardio Good Fitness. We looked to compare and determine characteristics of customers across different products.
We have been able to conclude that
1. Product with model number TM798 has more partnered high end customers with high income levels over 6000. High fitness levels and usage, averaging more than 4 times a week and higher miles expected to run. There is a much larger presence of males and average age for females is low in comparison at about 27 years.
2. Product with model number TM498 seems to be more of a generic product, maintaining just about average levels of income, fitness and usage across gender and marital status.
3. Usage and miles expected to run seems to be higher in females and partnered cutsomers for both these customers.
4. Product with model number TM195 is similar to TM498 but in contrast, has higher usage, fitness and miles expected to run for males and singles. It is slightly lesser in terms of income averages and has the most patronage.


## Recommendation

1. Product with model number TM798 should be targeted at customers expected to use at least 4 times a week, very particular about their fitness and with income averaging close to 6000. Particularly to partnered customers, older males with average age 29.5 and younger females around 27 years of age.
2. Product with model number TM498 should be targeted at relatively lower income earners and customers not expected to use too many times a week, less concerned about their fitness.
3. Product with model number TM195 should have similar targets to TM498 with lower income barrier more emphasis on fitter males and singles.
